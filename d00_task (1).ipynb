### 3.1 Объединение датасета
Вам нужно собрать полный датасет, собранный из частей датасета всех участников.
1. Датасет должен содержать в шапке названия полей.
2. Датасет не должен содержать лишних столбцов – только те, что были представлены в частях датасета.
3. Датасет должен быть отсортирован по возрастанию значений поля `ID`.
4. В датасете не должно быть дублирующихся строк и столбцов.

```python
import pandas as pd #мы будем часто пользоваться методами этой библиотеки, поэтому введем для нее сокращенное название – pd
import numpy as np

data1 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_01.csv')
data2 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_02.csv')
data3 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_03.csv')
data4 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_04.csv')
data5 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_05.csv')
data6 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_06.csv')
data7 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_07.csv')
data8 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_08.csv')
data9 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_09.csv')
data10 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_10.csv')
data11 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_11.csv')
data12 = pd.read_csv('/Users/npk-s80-1/Desktop/day00/dataset/part_12.csv')
data12.head()
```

```python
data_1_9 = pd.concat([data1, data9], ignore_index=True)
data_2_11 = pd.concat([data2, data11], ignore_index=True)
data_3_8 = pd.concat([data3, data8], ignore_index=True)
data_4_7 = pd.concat([data4, data7], ignore_index=True)
data_6_10 = pd.concat([data6, data10], ignore_index=True)
data_5_12 = pd.concat([data5, data12], ignore_index=True)
```

```python
data_itog = data_1_9.merge(data_2_11, how='left', sort=True, left_on='ID', right_on='ID')
data_itog = data_itog.merge(data_3_8, how='left', sort=True, left_on='ID', right_on='ID')
data_itog = data_itog.merge(data_4_7, how='left', sort=True, left_on='ID', right_on='ID')
data_itog = data_itog.merge(data_6_10, how='left', sort=True, left_on='ID', right_on='ID')
data_itog = data_itog.merge(data_5_12, how='left', sort=True, left_on='ID', right_on='ID')
data_itog.info(verbose=True)
```

### 3.2. Очистка текстовых данных
Теперь можно приступать к очистке данных. Для этого нужно:
1. Найти в получившемся датасете все поля, которые являются текстовыми.
2. Привести их значения к нижнему регистру.
3. В тех полях, где есть одновременно русское и английское написание одного и того же слова, оставить только русское (заменить английское на русское). Это условие не применять к полю `CLNT_JOB_POSITION`. Это будет чересчур даже для нашего курса :)
4. Если текстовое значение состоит только из “` `“ (пробел), то его заменить на `NaN` (пустое значение, а не строку 'NaN').
5. Если в текстовом значении есть цифры (телефон, дата рождения и др.), то всё значение текстового поля тоже заменить на `NaN`. Поле `PACK` оставить в исходном виде. Оно отвечает за пакет услуг клиента, и нам эта информация может быть важной. Затирать ее на пустые значения было бы опрометчиво.
6. Если есть столбцы, которые абсолютно идентичны по своим значениям для одних и тех же строк, то оставить только тот столбец, который идет первее. Речь идет только о текстовых полях.

```python
data_itog.APP_CAR = data_itog.APP_CAR.str.lower()
data_itog.APP_COMP_TYPE = data_itog.APP_COMP_TYPE.str.lower()
data_itog.APP_DRIVING_LICENSE = data_itog.APP_DRIVING_LICENSE.str.lower()
data_itog.APP_EDUCATION = data_itog.APP_EDUCATION.str.lower()
data_itog.APP_EMP_TYPE = data_itog.APP_EMP_TYPE.str.lower()
data_itog.APP_KIND_OF_PROP_HABITATION = data_itog.APP_KIND_OF_PROP_HABITATION.str.lower()
data_itog.APP_MARITAL_STATUS = data_itog.APP_MARITAL_STATUS.str.lower()
data_itog.APP_POSITION_TYPE = data_itog.APP_POSITION_TYPE.str.lower()
data_itog.APP_TRAVEL_PASS = data_itog.APP_TRAVEL_PASS.str.lower()
data_itog.CLNT_JOB_POSITION = data_itog.CLNT_JOB_POSITION.str.lower()
data_itog.CLNT_JOB_POSITION_TYPE = data_itog.CLNT_JOB_POSITION_TYPE.str.lower()
data_itog.CLNT_TRUST_RELATION = data_itog.CLNT_TRUST_RELATION.str.lower()
data_itog.PACK = data_itog.PACK.str.lower()
```
